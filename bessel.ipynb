```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt
from mpl_toolkits.axisartist.axislines import AxesZero
```

```python
plt.rcParams.update({
    'text.usetex': True,
    'font.family': 'serif',
})
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(axes_class=AxesZero)

ax.set_xlim((-100, 100))
ax.set_ylim((-0.5, 1.1))
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"${J}_n(x)$")
ax.grid(True)
fig.tight_layout()

for n in ["xzero", "yzero"]:
    ax.axis[n].set_axisline_style("-|>")
    ax.axis[n].set_visible(True)
for n in ["top", "right", "bottom", "left"]:
    ax.axis[n].set_visible(False)

x = np.linspace(-100, 100, 500)
for v in [-10]:
    ax.plot(x, sp.jv(v, x), label=r"${J}_{%s}$" % v)
ax.legend()
plt.show()
```
